```python
# Z score computation and outlier detection 
# z score describes position of raw score in terms of distance from mean , std deviavion . 
# an outlier is data point unusualy differs from other dateset (high or low )

# define columns to analyze 
target_cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

#initialize dataframe to store result 
outlier_report = pd.DataFrame(index=target_cols,columns=['Outlier_count','Outlice_pct'])
df_flags = df.copy() #create copy to store outliers 

for col in target_cols:
    if col in df.columns:
        z_score = np.abs(stats.zscore(df[col].dropna())) # compute z score for each column value Z = (X - ) /   
        df_flags[f'{col}_outlier']=False
        df_flags.loc[z_score > 3,f'{col}_outlier']=True

        outlier_count = df_flags[f'{col}_outlier'].sum()
        outlier_report.loc[col] =[
            outlier_count,f"{(outlier_count/len(df))*100:.2f}%"
        ]
print("Outlier Summary Report (|Z| > 3):")
print(outlier_report)

# Show rows with any outliers
print("\nSample of rows with outliers:")
print(df_flags[df_flags[[f'{col}_outlier' for col in target_cols if f'{col}_outlier' in df_flags]].any(axis=1)].head())
```
